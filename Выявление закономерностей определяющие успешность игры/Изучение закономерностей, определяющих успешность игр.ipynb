# Выявление закономерностей определяющие успешность игры.

# Описание проекта
Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

# Описание данных
<ul>
    <li><i>Name</i> — название игры</li>
    <li><i>Platform</i> — платформа</li>
    <li><i>Year_of_Release</i> — год выпуска</li>
    <li><i>Genre</i> — жанр игры</li>
    <li><i>NA_sales</i> — продажи в Северной Америке (миллионы проданных копий)</li>
    <li><i>EU_sales</i> — продажи в Европе (миллионы проданных копий)</li>
    <li><i>JP_sales</i> — продажи в Японии (миллионы проданных копий)</li>
    <li><i>Other_sales</i> — продажи в других странах (миллионы проданных копий)</li>
    <li><i>Critic_Score</i> — оценка критиков (максимум 100)</li>
    <li><i>User_Score</i> — оценка пользователей (максимум 10)</li>
    <li><i>Rating</i> — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг
        компьютерных игр и присваивает им подходящую возрастную категорию.</li>
</ul>

## Проведием исследовательский анализ данных 

```python
# функция для вывода простых гистограмм
def hist(df, x, y, bins, histfunc, title):
    fig = px.histogram(df,
                       x=x,
                       y=y,
                       nbins= bins,
                       histfunc = histfunc,
                       title = title)
    return fig
```

```python
#функция виолин с ящиком с усами
def box_plot (df, y, title):
    fig = px.violin(df,
                    y=y,
                    title = title,
                    box=True)
    return fig
```

```python
# функция для сравнивания гистограмм
def contrast_hist (df, x, y, color, histfunc, nbins):
    fig = px.histogram(df, 
                       x= x,
                       y= y, 
                       color= color, 
                       marginal= 'box', 
                       histfunc= histfunc, 
                       nbins= nbins, 
                       opacity=0.75,
                       barmode= 'overlay',
                       hover_data=df.columns)
    return fig
```

```python
#ящик с усами
def box (df, x, y):
    fig = px.box(df, x= x, y=y)
    return fig
```

```python
# диаграмма рассеяния
def bubble_charts(df, x, y, color):
    fig = px.scatter(df, x= x, y= y, color=color)
    return fig
```

## Проверим гипотезы

```python
#для того, что бы получить значение pvalue нам пришлось избавиться от NaN
#и так как рейтинги 0.0 нет на нужных нам платформах и жанрах мы выбрали его
actual_data['user_score'] = actual_data['user_score'].fillna(0.0)
```

```python
#создаем df для вывода гипотез по платформам
xbox_one_score = actual_data.query('(platform in "XOne") and (user_score != 0.0)')
pc_score = actual_data.query('platform in "PC" and (user_score != 0.0)')
#создаем df для вывода гипотез по жанрам
action_score = actual_data.query('genre in "Action" and (user_score != 0.0)')
sports_score = actual_data.query('genre in "Sports" and (user_score != 0.0)')
```

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H₀ всегда формулируется так, чтобы использовать знак равенства.

Мы определяем `нулевую гипотезу H₀`

**Нулевая гипотеза H₀** - средние пользовательские рейтинги платформ Xbox One и PC равны

**Альтернативная гипотеза H₁** - средние пользовательские рейтинги платформ Xbox One и PC не равны

```python
print('Дисперсия по выборке платформы Xbox One:' , np.var(xbox_one_score['user_score'] , ddof=1))
print('Дисперсия по выборке платформы PC:' , np.var(pc_score['user_score'] , ddof=1))
```

```python
print('средний пользовательский рейтинг платформы Xbox One: ', xbox_one_score['user_score'].mean())
print('средний пользовательский рейтинг платформы PC: ', pc_score['user_score'].mean())
print('медиана пользовательского рейтинга платформы Xbox One', xbox_one_score['user_score'].median())
print('медиана пользовательского рейтинга платформы PC', pc_score['user_score'].median())

alpha = 0.05

results =  st.ttest_ind(xbox_one_score['user_score'], pc_score['user_score'], equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

`Вывод:`  Они дейсвительно одинаковые
_______________________________________________________________________________________________________________________________

### Средние пользовательские рейтинги жанров Action и Sports разные

Мы определяем `нулевую гипотезу H₀`

**Нулевая гипотеза H₀** - средние пользовательские рейтинги жанров Action и Sports равны

**Альтернативная гипотеза H₁** - средние пользовательские рейтинги жанров Action и Sports не равны

```python
print('Дисперсия по выборке жанра Action:' , np.var(action_score['user_score'] , ddof=1))
print('Дисперсия по выборке жанра Sports:' , np.var(sports_score['user_score'] , ddof=1))
```

```python
print('средний пользовательский рейтинг жанра Action: ', action_score['user_score'].mean())
print('средний пользовательский рейтинг жанра Sports: ', sports_score['user_score'].mean())
print('медиана пользовательского рейтинга жанра Action', action_score['user_score'].median())
print('медиана пользовательского рейтинга жанра Sports', sports_score['user_score'].median())

alpha = 0.05

results =  st.ttest_ind(action_score['user_score'], sports_score['user_score'], equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

`Вывод:` Хоть среднее значения не сильно различается, но средние пользовательские рейтинги жанров Action и Sports действительно разные
