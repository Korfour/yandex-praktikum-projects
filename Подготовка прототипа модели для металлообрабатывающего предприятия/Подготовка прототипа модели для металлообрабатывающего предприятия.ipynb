### Эффективность обогащения

###### Проверим, что эффективность обогащения рассчитана правильно.
    
    Вычислим её на обучающей выборке для признака rougher.output.recovery.
    
    Найдём MAE между нашими расчётами и значением признака. 

Эффективность обогащения рассчитывается по формуле:

```python
def recovery (c, f, t):
    recovery = ((c*(f-t)) / (f*(c-t))) * 100
    return recovery
```

где:
C — доля золота в концентрате после флотации/очистки;

F — доля золота в сырье/концентрате до флотации/очистки;

T — доля золота в отвальных хвостах после флотации/очистки.

Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат.

```python
#recovery на обучающей выборке
train_recovery = recovery(gold_recovery_train['rougher.output.concentrate_au'],
                          gold_recovery_train['rougher.input.feed_au'],
                          gold_recovery_train['rougher.output.tail_au']
                         )
#значения наших вычислиней
train_recovery
```

```python
#данные значения
gold_recovery_train['rougher.output.recovery']
```

#### MAE между нашими расчётами и значением признака

```python
def mae (y, y1):
    mae = (y-y1).abs().sum() / len(y1)
    return mae
```

```python
mae(train_recovery, gold_recovery_train['rougher.output.recovery'])
```

Средняя абсолютная ошибка составляет 8e15 это указывает на то, что значения признаков в данных очень точны.

### Посмотрим, как меняется концентрация металлов (Au, Ag, Pb) на различных этапах очистки.

Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.

1. Флотация

Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.

На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).
2. Очистка

Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

### Исследуем суммарную концентрацию всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах.

```python
#сырьё
def raw_feed(df):
    return (df['rougher.input.feed_au'] + df['rougher.input.feed_ag']
            + df['rougher.input.feed_pb'] + df['rougher.input.feed_sol'])
```

```python
#черновой концентрат
def rougher_conc(df):
    return (df['rougher.output.concentrate_au'] + df['rougher.output.concentrate_ag']
            + df['rougher.output.concentrate_pb'] + df['rougher.output.concentrate_sol'])
```

```python
#финальный концентрат
def final_conc(df):
    return (df['final.output.concentrate_au'] + df['final.output.concentrate_ag']
            + df['final.output.concentrate_pb'] + df['final.output.concentrate_sol'])
```

```python
gold_recovery_full['rougher.input.feed'] = raw_feed(gold_recovery_full)
gold_recovery_full['rougher.output.concentrate'] = rougher_conc(gold_recovery_full)
gold_recovery_full['final.output.concentrate'] = final_conc(gold_recovery_full)
```

## Построим модель

### Напишем функцию для вычисления итоговой sMAPE.

```python
def final_smape(y, y1):
    y_rougher = y.iloc[:, 0]
    
    y1_rougher = y1[:, 0]
    #sMAPE rougher
    rougher_num = np.abs(y_rougher - y1_rougher)
    rougher_den = (np.abs(y_rougher) + np.abs(y1_rougher)) / 2
    smape_rougher = np.mean(rougher_num / rougher_den) * 100

    y_final = y.iloc[:, 1]
    y1_final = y1[:, 1]
    #sMAPE final
    final_num = np.abs(y_final - y1_final)
    final_den = (np.abs(y_final) + np.abs(y1_final)) / 2
    smape_final = np.mean(final_num / final_den) * 100
    #итоговое sMAPE
    final_smape = (smape_rougher * 0.25) + (smape_final * 0.75)
    return final_smape
```
